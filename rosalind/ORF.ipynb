## reverse complementary

```python
def reverse_complementary_strand(DNA_sequence):
    reverse_DNAseq = ""
    for base in DNA_sequence:
        if base == "A":
            reverse_DNAseq = "T"+ reverse_DNAseq
        if base == "C":
            reverse_DNAseq = "G" + reverse_DNAseq
        if base == "G":
            reverse_DNAseq = "C" + reverse_DNAseq
        if base == "T":
            reverse_DNAseq = "A" + reverse_DNAseq
    return reverse_DNAseq
```

```python
revDNAseq=reverse_complementary_strand(DNAseq)
```

```python
def translate_into_RNA(DNA_sequence):
    RNAseq = DNA_sequence.replace("T","U")
    return RNAseq
```

```python
def create_triplets(sequence, starting_position):
    codons = []
    for i in range(starting_position, len(sequence), 3): # the third argument to range is the step size!
        # start = i
        # end = start + 3
        # codon = s[start:end]
        codon = sequence[i:i+3] # both options are equally correct; this is just more compact!
        codons.append(codon)
    return codons
```

```python
#start codon position
def startcodons(triplets):
    start_codon_list=[]
    for i in range(len(triplets)):
        if triplets[i]=="AUG":
            #print(codons[i])
            start_codon_list.append(i)
    return start_codon_list
```

```python
starts0=startcodons(codons0)
print(starts0)

starts1=startcodons(codons1)
print(starts1)

starts2=startcodons(codons2)
print(starts2)

revstarts0=startcodons(revcodons0)
print(revstarts0)

revstarts1=startcodons(revcodons1)
print(revstarts1)

revstarts2=startcodons(revcodons2)
print(revstarts2)
```

```python
def stopcodons(triplets):
    stop_codon_list=[]
    for i in range(len(triplets)):
        if triplets[i]=="UAA" or triplets[i]=="UAG" or triplets[i]=="UGA":
            stop_codon_list.append(i)
    return stop_codon_list
```

```python
stops0=stopcodons(codons0)
print(stops0)

stops1=stopcodons(codons1)
print(stops1)

stops2=stopcodons(codons2)
print(stops2)

revstops0=stopcodons(revcodons0)
print(revstops0)

revstops1=stopcodons(revcodons1)
print(revstops1)

revstops2=stopcodons(revcodons2)
print(revstops2)
```

## Find ORF

```python
def find_orf(start, stop_codon_list,triplets):
    """
    smallest stop codon which is bigger then the start codon. 
    """ 
    for i in range(len(stop_codon_list)):
        if stop_codon_list[i] < start and start < stop_codon_list[i+1]:
            orf=triplets[start:stop_codon_list[i+1]]
            orf= "".join(orf)
            break
        elif stop_codon_list[i] > start:
            orf=triplets[start:stop_codon_list[i]]
            orf= "".join(orf)
            break
        #elif start > stop_codon_list[i]:
           # print("No ORF found")
            #continue
    return orf
```

```python
orf01=find_orf(8,stops0,codons0)
print(orf01)

orf02=find_orf(10,stops0,codons0)
print(orf02)

orf11=find_orf(1,stops1,codons1)
print(orf11)

revorf11=find_orf(23,revstops1,revcodons1)
print(revorf11)

revorf21=find_orf(1,revstops2,revcodons2)
print(revorf21)
```

## Translate into protein

```python
def translate_into_protein(orf):
    from util import codons
    peptide = ""
    for i in range(0, len(orf), 3):
        codon = orf[i:i+3]
        aminoacid = codons[codon]
        if aminoacid != "Stop":
            peptide = peptide + aminoacid
    return peptide
```

```python
print(translate_into_protein(orf01))
print(translate_into_protein(orf02))
print(translate_into_protein(orf11))
print(translate_into_protein(revorf11))
print(translate_into_protein(revorf21))
```

```python
def find_password(start, rotations_list):
    zeros = 0
    for r in rotations_list:
        if int(r[1:]) > 100:
            r = r[0] + str(int(r[1:]) - int(r[1]) * 100)
            zeros = zeros + int(r[1])
        if r[0] == "L":
            start = start - int(r[1:])
            if start < 0:
                start = 100 + start
            if start == 0:
                zeros = zeros +1
        elif r[0] == "R":
            start = start + int(r[1:])
            if start > 99:
                start = start - 100
            if start == 0:
                zeros = zeros +1
    return zeros
```

```python
password = find_password(50, rotations)
print(password)
```

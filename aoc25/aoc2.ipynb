# Check if string is invalid ID (so exactly what we need)

```python
def is_invalid(s):
    if len(s)%2 == 1:
        return False
    first_half = s[0: (len(s)//2)]
    second_half = s[len(s)//2:]
    if first_half == second_half:
        return True
    else: return False
```

# Make dictionary out of input data

```python
def create_ID_dictionary(allranges):
    IDranges = allranges.split(",")
    ID_dictionary={}
    for i in IDranges:
        i = i.split("-")
        ID_dictionary[i[0]] = i[1]
    return ID_dictionary
```

# Save all invalid IDs in a list

```python
def collect_all_invalid_IDs(ID_dictionary):
    invalidIDs = []
    for key, value in ID_dictionary.items():
        for number in range(int(key),int(value) +1):
            number = str(number)
            if is_invalid(number) == True:
                invalidIDs.append(number)
    return invalidIDs
```

# Putting everything in one function

```python
def invalid_IDs(allranges):
    dictionary_of_IDs = create_ID_dictionary(allranges)
    list_of_invalid_IDs= collect_all_invalid_IDs(dictionary_of_IDs)
    sum = 0
    for i in list_of_invalid_IDs:
        i = int(i)
        sum += i
    return sum
```
